# Explore Season Data
The purpose of this notebook is to explore the seasnal record data provided in the `records_history.csv` file and plan for data manipulation to be completed as part of pre-processing.

```python
# Import relevant libraries
import pandas as pd
```

```python
# Load end-of-seasson data
df = pd.read_csv('../data/records_history.csv')
df
```

```python
atr_df = df.groupby(by = ['Manager'])[['Wins', 'Losses', 'Ties']].sum()
atr_df['win_pct'] = (atr_df['Wins'] + 0.5*atr_df['Ties']) / (atr_df['Wins'] + atr_df['Losses'] + atr_df['Ties'])

atr_df.sort_values(by = 'win_pct', ascending=False)
```
